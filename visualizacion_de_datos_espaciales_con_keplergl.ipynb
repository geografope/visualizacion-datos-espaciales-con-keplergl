### **2. Llamado o activación de librerías**

```python
from google.colab import output
output.enable_custom_widget_manager()
from keplergl import KeplerGl
import pandas as pd
import geopandas as gpd
```

### **3. Kepler.gl con datos tabulares o csv con componente espacial**

```python
# Opendata: https://github.com/ambarja/snowfalldb-peru
url = 'https://raw.githubusercontent.com/ambarja/snowfalldb-peru/main/csv/snowfallDB-Peru.csv'
nevada_data = pd.read_csv(url)
```

```python
# Visualización de las 5 primeras filas
nevada_data.head()
```

```python
nevada_data = nevada_data[~nevada_data['Dep'].isin(['PIURA', 'LAMBAYEQUE','LIMA'])]
```

```python
# Visualización de datos tabulares (csv) con componenete espacial dentro de con kepler.gl
mapa_csv = KeplerGl(height=500, data={'nevada': nevada_data})
mapa_csv
```

### **4. Kepler.gl con geopandas**

```python
# Open data : https://github.com/ambarja/gpkg-pe
url_gpkg = 'https://github.com/ambarja/gpkg-pe/raw/main/prov_huancayo.gpkg'
gpkg = gpd.read_file(url_gpkg)
```

```python
# Visualización de las 5 primeras files
gpkg.head()
```

```python
# Visualización de geopackages dentro de con kepler.gl
mapa_gpkg = KeplerGl(height=500, data={"provincia-hcyo": gpkg})
mapa_gpkg
```

### **5. Configuración final del mapa web**

```python
config_mapa_csv = mapa_csv.config
config_mapa_gpkg = mapa_gpkg.config
```

### **6. Exportación del mapa web**

```python
mapa_csv.save_to_html(data={'nevada': nevada_data}, config=config_mapa_csv, file_name='mapa_csv.html')
mapa_gpkg.save_to_html(data={'provincia-hcyo': gpkg}, config=config_mapa_gpkg, file_name='mapa_gpkg.html')
```
